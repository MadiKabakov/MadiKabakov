```python
Sw = [0.30, 0.35, 0.40, 0.45, 0.50, 0.55, 0.60, 0.65, 0.70, 0.80]
Krw = [0.000, 0.005, 0.010, 0.017, 0.023, 0.034, 0.045, 0.064, 0.083, 0.120]
Kro = [1.000, 0.590, 0.320, 0.180, 0.080, 0.030, 0.010, 0.005, 0.001, 0.000]

data = pd.DataFrame({'Sw': Sw, 'Krw': Krw, 'Kro': Kro})

data.style.hide_index()
```

```python
#a)
Sorw = 0
for i in range(len(Sw)):
    if (Kro[i] == 0.0):
        Sorw = np.round(1 - Sw[i], 4)
print("Sorw: ", np.round(Sorw, 2))
#b)
Swc = 0
for i in range(len(Sw)):
    if (Krw[i] == 0.0):
        Swc = Sw[i]
print("Swc: ", np.round(Swc, 2))
#c)
KroSwc = 0
for i in range(len(Sw)):
    if(Sw[i] == Swc):
        KroSwc = Kro[i]

print("Kro(Swc): ", np.round(KroSwc, 4))
#d)
KrwSorw = 0
for i in range(len(Sw)):
    if (Sw[i] == 1 - Sorw):
        KrwSorw = Krw[i]


print("Krw(Sorw): ", np.round(KrwSorw, 4))
#e)
Muo = 1.30
Muw = 0.80

Kpodv = (KrwSorw / Muw) / (KroSwc / Muo)
print("коэффициент подвижности для воды: ", Kpodv )
```
