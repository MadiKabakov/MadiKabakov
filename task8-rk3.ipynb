```python
def fVp(F, C):
    return 5.81 - 9.41 * F - 2.21 * C
    
def fVs(F, C):
    return 3.89 - 7.07 * F - 2.04 * C

F = 0.2
C = 0.05

Vp3 = fVp(F, C)
Vs3 = fVs(F, C)

print("Vp: ", np.round(Vp3, 4), "км/с")
print("Vs: ", np.round(Vs3, 4), "км/с")
print("отношение Vp / Vs: ", np.round(Vp3 / Vs3, 4))


```
