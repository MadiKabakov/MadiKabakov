```python
#4 задание
data4 = pd.DataFrame({'Слой': [1, 2, 3], \
                      'Проницаемость (мД)': [126, 4, 35]})
data4.style.hide_index()
```

```python
#а)
mD1 = data4['Проницаемость (мД)']
sum = 0
for i in range(len(mD1)):
    sum += mD1[i]
srSUM = sum / len(mD1)
print("Среднеарифметическая будет: ", srSUM, " мД")

#б)
sum1 = 0
for i in range(len(mD1)):
    sum1 += 1 / mD1[i]
garSUM = sum1 / len(mD1)
print("Среднегармоническая будет: ", np.round(1 / garSUM, 4), " мД")

#в)

Vdp = 1 - math.exp(-math.sqrt(math.log(srSUM * garSUM)))
print("коэффициент Дикстра-Парсонса: ", np.round(Vdp, 4))
```
